# American Airlines tickets analysis

#### Let's analyze the dynamics of changes in prices for airline tickets of American airlines

#### Our data we will gets from csv file which include: Date - date in days, Open - price from the first transaction of a trading day, High - maximum price in a trading day, Low - minimum price in a trading day, Clode - price from the last transaction of a trading day, Adj_Close - closing price adjusted to reflect the value after accounting for any corporate actions, Volume - number of units traded in a day.

### Import lib

```python
import pandas as pd

import matplotlib.pyplot as plt
```

### Import our data from csv file

```python
data = pd.read_csv("C:/Users/User/OneDrive/Рабочий стол/Типа Analytic/Project/AAL.csv", parse_dates=True)

data.head()
```

## For analysis we take Adj_Close and make analys by year

### Make a new column Year

```python
data['Year'] = pd.DatetimeIndex(data['Date']).year

data.head()
```

### Group our date by Year

```python
res = data.groupby('Year').mean('Adj_Close')['Adj_Close']

res.head(18)
```

### Make a plot Adj_Close by Year

```python
res.plot.bar(xlabel='Year', ylabel = 'USD')

plt.gcf().set_size_inches(20, 5)

plt.show()
```

#### We can see Adj_Close have jump recessions from 2007 to 2009 and from 2019 to 2020

### Let's make our plot like line by dates

```python
res1 = data[['Date','Adj_Close']]

res1 = res1.set_index('Date')

res1.plot(xlabel='Date', ylabel='USD')

plt.gcf().set_size_inches(20, 5)

plt.show()
```

### After analyzing the economic situation in the years of a strong recession, we can draw the following conclusions:
### 1) decline from 2007 to 2009 due to the financial crisis
### 2) decline fromm 2019 to 2020 due to the pandemic
